Термин ***система*** будет неоднократно употребляться в дальнейшем при изложении теории марковских процессов. Поэтому вначале сделаем небольшое отступление и дадим определение этого понятия. 

***Системой*** называется целенаправленное множество взаимосвязанных элементов любой природы.
 
Более строгое математически определение системы такое:

Пусть A = {a_i} – множество некоторых объектов, обладающих определенными свойствами (элементы системы); R = {r_i} – связи между указанными объектами; Z – цель функционирования системы. Тогда систему можно определить, как совокупность элементов со связями и целью функционирования системы, т.е. S = {A, R, Z}

Различают два вида систем: 
1. Детерминированные;
2. Стахостатические.

***Детерминированными системами*** называются такие системы, состояние которых однозначно определяется начальными значениями и может быть однозначно предсказано для любого момента времени. Например, движение планет в солнечной системе - они всегда движутся по заранее известным законам физики, дверной замок - после вращения ключа мы всегда знаем в какую сторону будет двигаться ригель. Проще говоря, поведение нашей системы в любой момент времени **предсказуемо**. 

***Стохастические системы*** – это такие системы, эволюция которых во времени не может быть описана однозначно. То есть будущее поведение системы и прогноз ее поведения носят случайный характер. Например, прогноз погоды - мы не можем со 100% уверенностью утверждать, что через несколько дней будет солнечная безветренная погода, так как всегда есть угроза дождя или произойдет изменение интенсивности солнечного излучения. Нашего знания о текущем состоянии системы и особенностей связей между ее составными элементами недостаточно для однозначного предсказания будущего системы. 

Поведение некоторых стохастических систем,  мы можем моделировать с помощью марковских случайных процессов, к обсуждению которых мы и переходим.

# Марковские процессы

Случайный процесс называется марковским, если ***для любого момента времени вероятностные характеристики процесса в будущем
зависят только от его состояния в данный момент времени и не зависят от того, когда и как система пришла в это состояние.*** («Будущее
зависит от прошлого только через настоящее»).

## Классификация марковских процессов 
Все марковские процессы можно разделить на классы в зависимости от:
1. Множество состояний системы {S}; 
2. Множество моментов времени наблюдения {T}.

Если множество {S} дискретное, то случайный процесс называется ***цепью Маркова***.

Когда 
1. {T} − дискретное множество, то процесс называется ***цепью Маркова с дискретным временем*** (или ***дискретной цепью Маркова***);
2. {T} − непрерывное множество, то процесс называется ***цепью Маркова с непрерывным временем*** (или ***непрерывной цепью Маркова***).

Если {S} непрерывное и {T} непрерывное, то процесс называется ***непрерывным марковским процессом***.

Далее будут рассматриваться **цепи Маркова с дискретным и непрерывным временем**. 

# Цепи Маркова с дискретным временем

Рассмотрим случайный процесс, при котором система может находиться в состояниях S_1, S_2, ..., S_k, а изменения состояний могут происходить в определённые моменты времени t_1, t_2, ..., t_n.

Пусть в момент времени t(m), система находилась в состоянии S_i, a в момент времени t(m+1) в состоянии S_j.

Переходы из одного состояния в другое происходят с вероятностью p_i_j , через равные промежутки времени |t(m+1) - t(m)| = 1, которые мы будем называть ***шагом***.

Процесс будет марковским, если эта вероятность на каждом шаге зависит только от состояния, в которое система попала на предыдущем шаге, и не зависит от предыдущих шагов. Иначе говоря, для марковского процесса переходная вероятность не должна зависеть от m.

Переходный вероятности образуют ***матрицу переходных вероятностей***.

Цепи маркова с дискретным временем бывают ***однородные*** и ***неоднородные***.

Если вероятности переходов не зависят от момента времени t , то есть p_i_j(t) = p_i_j , то цепь Маркова называется ***однородной*** с матрицей вероятностей переходов за один шаг:

![](pictures\possibilities_matrix.png)

Для ***однородной цепи Маркова*** стохастическая матрица P имеет постоянный вид, для любого шага, т.е. P(1) = P(2) = P(3) = ... P(n), где (1), (2), ... (n) – номера шагов.

Если исследуемая ***цепь Маркова неоднородная***, то матрицы вероятностей переходов для
каждого шага P(1), P(2), P(3), ..., P(n) будут разными. 

Для того чтобы определить цепь Маркова с дискретным временем необходимо обязательно задать ***начальное распределение***.
Т.е. начальное распределение – это распределение вероятностей состояний системы в момент времени t = 0 . Этим моментом может быть например тот, начиная с которого мы стали наблюдать за нашей системой.

# Стационарное распределение вероятностей состояний системы 

Распределение вероятностей называется ***стационарным***, если оно не изменяется от шага к шагу.

Для того, чтобы марковская цепь с конечным числом состояний имела единственное стационарное предельное распределение вероятностей, необходимо и достаточно, чтобы все существенные состояния были сообщающимися.

Состояние называется сообщающимся, если из него можно перейти в другие состояния.

***Предельная стационарная вероятность несущественного состояния всегда оказывается равной нулю***: рано или поздно система выходит из этого состояния и уже в него не возвращается. 

***Предельная стационарная вероятность единственного состояния поглощения (при отсутствии других классов существенных состояний) всегда оказывается равной единице***: рано или поздно система придёт в это состояние и уже из него не выйдет.

**Задание 2:**

Матрица парехода P имеет вид: 

| 0.4 | 0.6 |

| 0.3 | 0.7 |

Начальное распределение вероятностей: (0.1, 0.9)

Найти:
1. матрицу перехода за 2 шага;
2. распределение вероятностей по состояниям после 2-го шага;
3. вероятность того, что после 1-го шага состоянием цепи будет S_1
4. стационарное распределение вероятностей по состояниям. 

```python
import numpy as np


# Запишем начальное распределение
initial_distribution = np.array([0.1, 0.9])

# Запишем матрицу переходных вероятностей
P = np.array([[0.4, 0.6] , [0.3, 0.7]])

# 1: Матрица перехода за 2 шага
P_2 = np.linalg.matrix_power(P, 2)
print(f'Матрица переходов за 2 шага: \n{P_2}')
# Как мы видим сумма по строкам осталась равна 1 (самопроверка)

# 2: Распределение вероятностей по состояниям после 2-го шага
distribution_2 = np.matmul(initial_distribution, P_2)
print(f'Распределение вероятностей по состояниям после 2-го шага: \n{distribution_2}')
# Как мы видим сумма осталась равна 1 (самопроверка)

# 3: Вероятность того, что после 1-го шага состоянием цепи будет S_1
# Для этого нам нужно перемножить начальное распределение на первый стобец матрицы преходных вероятностей
P1 = np.matmul(initial_distribution, P[:, :1])
print(f'Вероятность того, что после 1-го шага состоянием цепи будет S_1: \n{P1}')
```

```python
# 4: Стационарное распределение вероятностей по состояниям
############
# СПОСОБ 1 #
############
# Будем возводить нашу исходную матрицу переходных вероятностей
# до тех пор пока распределение станет стационарным
for i in range(0, 10, 2):
    print('-' * 100)

    matrix = np.linalg.matrix_power(P, i)
    distribution = np.matmul(initial_distribution, matrix)
    print(f'Шаг = {i}\n', distribution)
```

```python
# Уже на 2 шаге мы получили стационарное распределение по вероятностям
# Выведем их в виде дробей (Fraction)
from fractions import Fraction

print([Fraction(number).limit_denominator(1000) for number in distribution])
```

Смысл полученного результата таков: в установившемся режиме система будет одну треть времени проводить в состоянии S_1 и две трети — в состоянии S_2.

```python
# Будем считать, что система S (данное предприятие) находится в
# состоянии S1, если выпускается товар А, и в состоянии S2 , если выпускается
# товар B. Если система находится в состоянии S1 , то с вероятностью 0,5 она
# останется в этом состоянии к концу недели и с вероятностью 0,5 может перейти в состояние S2.
# Если же система находится в состоянии S2, то выпускается товар В, и с вероятностью 0,4 система может вернуться в состояние
# S1, а с вероятностью 0,6 может остаться в состоянии S2. Таким образом,
# матрица вероятности переходов имеет вид

P = np.array([[0.5, 0.5], [0.4, 0.6]])

initial_distribution = np.array([1, 0])

# Определим распределение вероятностей состояний системы через 1 шаг (через неделю).
distribution_1 = np.matmul(initial_distribution, P)
print(f'C вероятностью = {P1[0]} предприятие будет выпускать товар А через неделю')

# Определим распределение вероятностей состояний системы через 2 шаг (через 2 недели).
distribution_2 = np.matmul(initial_distribution, np.linalg.matrix_power(P, 2))
print(f'C вероятностью = {distribution_2[0]} предприятие будет выпускать товар А через неделю')

# Определим стационарные вероятности
for i in range(0, 10, 2):
    print('-' * 100)

    matrix = np.linalg.matrix_power(P, i)
    distribution = np.matmul(initial_distribution, matrix)
    print(f'Шаг = {i}\n', distribution)

print(f'Стационарное распределение вероятностей: {[Fraction(number).limit_denominator(1000) for number in distribution]}')
```

# Цепь Маркова с непрерывным временем

Ранее мы рассмотрели дискретные цепи Маркова, когда система может находиться в одном из дискретных состояний S_1, S2, ..., S_n и в **фиксированные** моменты времени t_1, t_2, ..., t_n переходит из одного состояния в другое.

Рассмотрим теперь марковский процесс конечным или счетным множеством состояний {S}, который изменяет свои состояния в случайные моменты времени. Такой процесс
называется ***цепью Маркова с непрерывным временем***. 

***В этом случае говорить о вероятности перехода системы из одного состояния в другое в любой определённый момент времени бессмысленно*** (она равна нулю, подобно вероятности любого значения непрерывной случайной величины). ***Можно говорить только о вероятности перехода за промежуток времени ∆t (подразумевается, что за малый промежуток изменение не может произойти более 1 раза).***

Поэтому мы имеем дело с плотностью вероятности. А для цепи маркова с непрерывным временем плотоность вероятности перехода системы из состояния S_i в состояние S_j называется  ***интенсивностью перехода***.

***Поскольку интенсивность можно рассматривать как производную от вероятности перехода, её знак показывает направление изменения вероятности этого перехода с ростом времени.*** Вероятность перехода в **иное** состояние, естественно, растёт. Вероятность же остаться **в том же состоянии со временем падает.**

**Задание 3:**

Матрица интесивностей имеет вид: 

| 0 | 0 | 1 | 3 |

| 4 | 0 | 3 | 0 |

| 0 | 3 | 0 | 4 |

| 0 | 1 | 1 | 0 |

- Построить граф переходов между состояниями, ребра которого помечены соответствующими интенсивностями переходов.
- Найти стационарное распределение вероятностей

```python
intensity_matrix = np.array([[0, 0, 1, 0],
                             [4, 0, 3, 0],
                             [0, 3, 0, 4],
                             [0, 1, 1, 0]])

# Создадим граф на основе переходных вероятностей
sources, targets = np.where(intensity_matrix)
weights = intensity_matrix[sources, targets]
edges = list(zip(sources, targets))
edge_labels = dict(zip(edges, weights))

# Нарисуем граф
fig, ax = plt.subplots(figsize=(15, 7))
Graph(edges, edge_labels=edge_labels, edge_label_position=0.66, node_labels=True, arrows=True, ax=ax, width=1)
plt.show()
```

```python
import numpy as np
from fractions import Fraction


def get_equation(matrix: np.ndarray, indx: int) -> np.array:
    """
    Получение i-го вектора для уравнений Колмогорова

    Args:
        matrix (np.ndarray): Матрица Марквоского процесса
        indx (int): Индекс вектора в уравнениях Колмагорова

    Returns:
        np.array: Полученный вектор
    """
    vector = [0 for _ in range(len(matrix) + 1)]

    input_intesity = matrix[:, indx]
    vector[0: len(matrix)] = input_intesity

    output_intensity = sum(matrix[indx, :]) * -1
    vector[indx] = output_intensity

    return vector

intensity_matrix = np.array([[0, 0, 1, 0],
                   [4, 0, 3, 0],
                   [0, 3, 0, 4],
                   [0, 1, 1, 0]])
system_of_equations = np.array([[1 for _ in range(len(intensity_matrix) + 1)]])

for indx in range(len(intensity_matrix)):
    vector = get_equation(intensity_matrix, indx)
    system_of_equations = np.vstack([system_of_equations, vector])

print(f'Система уравнений Колмогорова \n {system_of_equations}')

# Решим систему
system_of_equations = system_of_equations[:-1, :]
a = system_of_equations[:, :-1]
b = system_of_equations[:, -1]
result = np.linalg.solve(a, b)
result = [Fraction(number).limit_denominator(1000) for number in result]
print(f'Стационарные вероятности: {result}')
```

```python
import numpy as np
from fractions import Fraction


def get_equation(matrix: np.ndarray, indx: int) -> np.array:
    """
    Получение i-го вектора для уравнений Колмогорова

    Args:
        matrix (np.ndarray): Матрица Марквоского процесса
        indx (int): Индекс вектора в уравнениях Колмагорова

    Returns:
        np.array: Полученный вектор
    """
    vector = [0 for _ in range(len(matrix) + 1)]

    input_intensity = matrix[:, indx]
    vector[0: len(matrix)] = input_intensity

    output_intensity = sum(matrix[indx, :]) * -1
    vector[indx] = output_intensity

    return vector

intensity_matrix = np.array([[0, 1, 0, 0],
                   [1, 0, 0, 1],
                   [0, 2, 0, 1],
                   [0, 0, 3, 0]])
system_of_equations = np.array([[1 for _ in range(len(intensity_matrix) + 1)]])

for indx in range(len(intensity_matrix)):
    vector = get_equation(intensity_matrix, indx)
    system_of_equations = np.vstack([system_of_equations, vector])

# Решим систему
system_of_equations = system_of_equations[:-1, :]
a = system_of_equations[:, :-1]
b = system_of_equations[:, -1]
result = np.linalg.solve(a, b)
result = [Fraction(number).limit_denominator(1000) for number in result]
print(f'Стационарные вероятности: {result}')

# Найдем матожидание
E = sum([indx * value  for indx, value in enumerate(result)])
print(f'Мат. ожидание: {E}')
```

# Системы массового обслуживания

***Системами массового обслуживания*** называются системы, удовлетворяющие массовый спрос в форме так называемых заявок, поступающих
в случайные моменты времени и образующих ***входящий поток заявок***. Обслуженные заявки образуют ***выходящий поток обслуженных заявок***. Примеры СМО: телефонные станции, ремонтные мастерские, билетные кассы, магазины и т.д.

Устройство, обслуживающее заявку, называется ***каналом обслуживания***. СМО может быть ***одноканальной*** или ***многоканальной***. Примеры каналов: линии связи, лифты, продавцы и т.д.

Если заявка, поступившая в момент занятости СМО, покидает её без обслуживания, то СМО называется СМО ***с отказами***. Если же заявка ожидает освобождения канала обслуживания, то это СМО ***с очередью*** (с ожиданием).

Процесс работы СМО представляет собой ***случайный процесс с дискретными состояниями и непрерывным временем (цепь Маркова).***

# Многоканальная СМО с отказами

**Задача Эрланга.** Имеется n каналов, на которые поступает поток заявок с интенсивностью λ. Поток обслуживаний имеет интенсивность μ.

Будем рассматривать систему, работающую без очереди, т.е. с отказами в обслуживании заявкам, пришедшим в тот момент, 
когда все каналы системы заняты обслуживанием ранее пришедших заявок.

Состояния системы будем нумеровать по числу заявок, находящихся на обслуживании в системе, т. е. S_k — состояние системы, когда в ней
находится k заявок (занято k каналов).

Граф системы выглядит следующим образом: 

![](pictures\smo_more_channels.png)

Поток заявок в систему не зависит от состояния каналов системы, поэтому при любом числе занятых каналов этот поток сохраняет свою интенсивность, 
следовательно, для любого состояния системы входящий в это состояние поток заявок (это поток событий, переводящих систему из предыдущего состояния в данное) 
имеет интенсивность λ. Интенсивность потока обслуживаний равна μ. Это значит, что каждый канал может обслуживать непрерывный поток заявок с интенсивностью μ. 
Событием такого потока является выход обслуженной заявки из системы, т. е. освобождение канала. 
Следовательно, μ можно рассматривать как интенсивность освобождения канала после очередного обслуживания заявки. 
Если в системе заняты два канала (т. е. ее состояние S_2 ), то может освободиться любой из них с интенсивностью μ, это значит, 
что суммарная интенсивность перехода в состояние S1 (занят один канал) равна 2μ. Точно так же, если система находится в состоянии Sk (заняты k каналов), 
то интенсивность входящего в такое состояние системы потока заявок равна λ, а интенсивность выходящего из этого состояния системы потока обслуживаний 
(т.е. интенсивность освобождения одного из k занятых каналов) равна kμ, т.к. любой из занятых k каналов может освободиться с интенсивностью μ. 
Таким образом, kμ — это интенсивность перехода системы из состояния Sk в состояние Sk−1 . 
При этом λ — это интенсивность перехода системы под влиянием потока заявок из состояния Sk−1 в Sk .

Введем новый параметр: **p = λ / μ**

Этот параметр называют приведенной интенсивностью потока заявок, или интенсивностью нагрузки каналов, или трафиком. 
Он выражает среднее число заявок, приходящих за среднее время обслуживания одной заявки.

**Задание 5:**

Определить минимально допустимое количество операторов call-центра, если известно, что в среднем, поступает 1,6 звонка в

```python
# Выпишем начальные параметры
λ = 1.6
μ = round(1 * 60 / 88, 2)

p = round(λ / μ, 2)

print(f'Интенсивность поступления звонков {λ = } (мин^(-1))')
print(f'Время обработки одной заявки {μ = } (мин^(-1))')
print(f'Коэффициент загруженности СМО {p = }')
```

```python
# Для того, чтобы количество необслуженных клиентов составляло не более 10%, необходимо чтобы параметр Q был больше 0.9 или p_отк меньше 0.1
# Для этого нам необходимо увеличивать количество каналов

from math import factorial

def calculate_p_0(p: float, n: int) -> float:
    """
    Расчет вероятности нахождения системы в состоянии S_0
    (все каналы свободны)

    Args:
        p (float): Коэффициент загруженности СМО
        n (int): Количество каналов

    Returns:
        Рассчитанная вероятность
    """

    p_0 = 0
    for i in range(n + 1):
        p_0 += p ** i / factorial(i)

    p_0 = round(pow(p_0, -1), 4)
    return p_0

def calculate_p_n(p: float, n: int) -> float:
    """
    Расчет вероятности отказа обслуживания
    (все каналы заняты или система в состоянии S_n)

    Args:
        p (float): Коэффициент загруженности СМО
        n (int): Количество каналов

    Returns:
        Рассчитанная вероятность
    """

    p_0 = calculate_p_0(p, n)

    p_n = p ** n * p_0 / factorial(n)
    p_n = round(p_n, 4)
    return p_n

def calculate_Q(p: float, n: int) -> float:
    """
    Расчет вероятности  обслуживания заявки

    Args:
        p (float): Коэффициент загруженности СМО
        n (int): Количество каналов

    Returns:
        Рассчитанная вероятность
    """
    p_n = calculate_p_n(p, n)
    Q = round(1 - p_n, 4)
    return Q

def calculate_A(p: float, n: int, λ: float) -> float:
    """
    Расчет вероятности  обслуживания заявки

    Args:
        p (float): Коэффициент загруженности СМО
        n (int): Количество каналов
        λ (float): Интенсивность поступления заявок

    Returns:
        Рассчитанная вероятность
    """
    Q = calculate_Q(p, n)
    A = round(Q * λ, 4)
    return A
```
